```python
data1.replace('No internet service',"No",inplace=True)
data1.replace('No phone service',"No",inplace=True)
```

```python
def print_unique_values(data1):
    for col in data1: 
            print(col,": ",data1[col].unique())
```

```python
yes_no_cols = ['Partner',"Dependents","PhoneService",'MultipleLines',"OnlineSecurity","OnlineBackup","DeviceProtection",
              'TechSupport',"StreamingTV","StreamingMovies",'PaperlessBilling',"Churn",""]
```

```python
for col in yes_no_cols:
    data1[col].replace({
        "Yes": 1, 
        "No": 0
    },inplace=True)
```

```python
print_unique_values(data1)
```

```python
data1.gender.replace({"Female":0,"Male":1},inplace=True)
```

```python
print_unique_values(data1)
```

```python
data2 = pd.get_dummies(data1)
```

```python
data2.head()
```

# Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

cols_to_scale  = ['tenure',"TotalCharges","MonthlyCharges"]


data2[cols_to_scale] =  scaler.fit_transform(data2[cols_to_scale])
```

```python
data2["TotalCharges"]
```

```python
X = data2.drop("Churn",axis=1)
y = data2["Churn"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test , y_train,y_test = train_test_split(X,y, test_size=0.2,random_state = 5)
```

```python
X_train.shape
```

```python
import tensorflow as tf 
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,),activation="relu"),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dense(10,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
    
])

model.compile(optimizer = "adam",loss="binary_crossentropy",metrics=["accuracy"])

model.fit(X_train,y_train,epochs=25)
```

```python
model.evaluate(X_test,y_test)
```

```python
predictions = model.predict(X_test)
predictions[:5]
```

# Convertir le tableau en 0 et 1 

Comparer chaque valeure si <0.5  alors c'est 0 si non c'est 1

```python
pred = [] 
for el in predictions : 
    if el> 0.5: 
        pred.append(1)
    else:
        pred.append(0)
        
```

```python
pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,pred))
```

```python
import seaborn as sns

cm = tf.math.confusion_matrix(labels=y_test,predictions = pred)

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt="d")

plt.xlabel('Predictions')
plt.ylabel("Truth")
```
